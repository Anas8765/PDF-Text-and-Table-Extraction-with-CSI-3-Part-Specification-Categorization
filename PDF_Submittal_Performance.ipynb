```python
import tabula
from PyPDF2 import PdfReader

# Creating a PDF reader object
reader = PdfReader('/content/PDF 1 - Architects Spec.pdf')

# List to store text from all pages
all_text = []

# List to store tabular data
tables = []

# Iterating over all pages and extracting text
for page_number, page in enumerate(reader.pages, start=1):
    # Extracting text from page
    text = page.extract_text()
    all_text.append(text)

    # Extracting tables from page using Tabula
    df = tabula.read_pdf('/content/PDF 1 - Architects Spec.pdf', pages=page_number, multiple_tables=True)
    tables.extend(df)

# Concatenating text from all pages into a single string
full_text = "\n".join(all_text)

# Printing extracted text
print(full_text)

# Printing tabular data
for table_number, table in enumerate(tables, start=1):
    print(f"\nTable {table_number}:\n{table}")
```

```python

example = '''You are an expert in making chart for CSI 3-part specifications of only submittal and product requirements based on the text given by the user. Do not write anything besides submittal and product requirements.

In determining and categorizing requirements into submittal and product the definition of them are as follows:

1. Submittal Requirements:
    - Submittal requirements refer to the documents or information that a contractor or supplier must provide to the project owner, architect, engineer, or other relevant parties for review and approval before proceeding with the work.
    - These requirements typically include documents such as product data, shop drawings, samples, test reports, product certificates, energy performance certificates, qualification data, mockups, sample warranties, sustainable design submittals and any other relevant information specified in the contract documents.
    - Submittal requirements ensure that the materials, products, and work methods proposed by the contractor meet the project specifications, codes, standards, and other regulatory requirements.
2. Product Requirements:
    - Product requirements, on the other hand, refer to the specifications, standards, and performance criteria that a particular product or material must meet to be considered acceptable for use in the project.
    - These requirements are typically outlined in the project specifications or other contract documents and may include criteria such as material composition, strength, durability, dimensions, fire performance, structural performance, water penetration, materials used, fabrication, finishes, and other relevant characteristics.
    - Product requirements ensure that the materials and products used in the project meet the necessary quality and performance standards to achieve the desired outcomes and comply with applicable regulations.

Apply these reasonings to the following text.
'''

prompt = f'''{example} {full_text}'''

completion = co.chat(
  message=prompt,
  model='command-r',
  temperature=0.2,
)

print(completion.text)
```

Conversion to table

```python
import pandas as pd

lines = completion.text.strip().split('\n')

# Initialize variables to store data
category = None
data = []

# Loop through lines and extract data
for line in lines:
    if line.startswith("Submittal Requirements:"):
        category = "Submittal Requirements"
    elif line.startswith("Product Requirements:"):
        category = "Product Requirements"
    elif line.startswith("-"):
        requirement = line.split(": ", 1)[0][2:].strip()
        description = line.split(": ", 1)[1].strip()
        data.append([category, requirement, description])

# Create a DataFrame
df = pd.DataFrame(data, columns=["Category", "Requirements", "Description"])

# Print the DataFrame
df
```
